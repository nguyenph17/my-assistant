```python
from duckduckgo_search import AsyncDDGS

query = "Mã cổ phiếu của Công ty hoá chất Đứa Giang là gì?"
results = await AsyncDDGS().achat(query, model="gpt-4o-mini")
print(results)
```

```python
from langchain_community.document_loaders import AsyncHtmlLoader

urls = ["https://cafef.vn/gan-5000-ty-dong-trai-phieu-chay-ve-nhom-cong-ty-bat-dong-san-trong-thang-8-2024-18824090720383859.chn"]
loader = AsyncHtmlLoader(urls)
docs = loader.load()
```
